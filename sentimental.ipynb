```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential  # <-- Capitalized 'S'
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
```

```python
vocab_size=10000
(x_train,y_train),(x_test,y_test)=imdb.load_data(
    num_words=vocab_size
)
print("Training sample:",len(x_train))
print("TEsting Samples:",len(x_test))
```

```python
max_length=200
x_train = pad_sequences(
    x_train,
    maxlen=max_length,
    padding='post'
)

x_test = pad_sequences(
    x_test,
    maxlen=max_length,
    padding='post'
)
```

```python
model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=64, input_length=max_length),
    SimpleRNN(64),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
model.summary()
```

```python
history = model.fit(x_train, y_train, epochs=5, batch_size=64, validation_split=0.2)
```

```python
loss,accuracy=model.evaluate(x_test,y_test)
print('Accuracy',accuracy)
```

```python
from tensorflow.keras.datasets import imdb

# Get the word index dictionary
word_index = imdb.get_word_index()

# Crucial fix: changed .item() to .items()
word_index = {k: (v + 3) for k, v in word_index.items()}

# Set up the special tokens
word_index["<PAD>"] = 0
word_index["<START>"] = 1
word_index["<UNK>"] = 2
word_index["<UNUSED>"] = 3
```

```python
def review_to_sequence(review):
  review=review.lower().split()
  sequence=[]
  for word in review:
    sequence.append(word_index.get(word,2))
  return sequence
```

```python
from string import printable
max_length =200
def predict_sentiment(review):
  sequence=review_to_sequence(review)
  padded=pad_sequences(
      [sequence],
      maxlen=max_length,
      padding='post'
  )
  prediction=model.predict(padded,verbose=0)
  score = prediction[0][0]
  print("sentiment Score:",score)
  if score>0.5:
    print("postive Review")
  else:
    print("Negtive Review")
```

```python
predict_sentiment(
    "movie was ulter flop"
)
```
