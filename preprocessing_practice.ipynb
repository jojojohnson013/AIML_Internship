```python
df['Age_mean_inputed']=df['Age'].fillna(df['Age'].mean())
df['Age_median_inputed']=df['Age'].fillna(df['Age'].median())
```

```python
df['Education_mode_inputed']=df['Education'].fillna(df['Education'].mode()[0])
df['Monthly_Salary_mean_inputed']=df['Monthly_Salary'].fillna(df['Monthly_Salary'].mean())
df['Education_new']=df['Education'].fillna(df['Education'].mode()[0])
df['city_mode_inputed']=df['City'].fillna(df['City'].mode()[0])
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler,RobustScaler

file_path="/content/preprocessing_practice_dataset.csv" # Corrected file path
prepro_data=pd.read_csv(file_path) # Changed titanic_data to prepro_data

numerical_columns=['Age','Performance_Rating'] # Adjusted columns to match actual data
prepro_data.dropna(subset=numerical_columns,inplace=True)

minmax_scale=MinMaxScaler()
standard_scaler=StandardScaler()
robust_scaler=RobustScaler()

prepro_data['Age_MinMAx']=minmax_scale.fit_transform(prepro_data[['Age']]) # Changed titanic_data to prepro_data and standard_scaler to minmax_scale
prepro_data['Performance_Rating_MinMax']=minmax_scale.fit_transform(prepro_data[['Performance_Rating']]) # Changed Fare to Performance_Rating and standard_scaler to minmax_scale

prepro_data['Age_Standard']=standard_scaler.fit_transform(prepro_data[['Age']])
prepro_data['Performance_Rating_Standard']=standard_scaler.fit_transform(prepro_data[['Performance_Rating']]) # Changed Fare to Performance_Rating

prepro_data['Age_Robust']=robust_scaler.fit_transform(prepro_data[['Age']])

print("DataFrame with MinMax and Standard Scaled columns (first 5 rows):")
display(prepro_data.head())
```
